# LOAD DATA

```python


df = pd.read_csv("../data/Grammar_Correction.csv")
df.columns = df.columns.str.strip()

input_texts = df['Ungrammatical Statement'].astype(str)
target_texts = df['Standard English'].astype(str)
```

# DARASET

```python
pairs = []

for inp, out in zip(input_texts, target_texts):
    inp_words = inp.split()
    out_words = out.split()
    
    for i in range(min(len(inp_words), len(out_words))):
        pairs.append((inp_words[i], out_words[i]))
```
